````python
# Function for ploting the decision boundaries of a model
def plot_decision_regions(X, y, classifier, resolution=0.01, title=""):

    # setup marker generator and color map
    markers = ('.', '.')
    colors = ('blue', 'red')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.title(title)
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')
    plt.show()
````

Use 5-fold cross validation in order to find the best eps and eta params from the given lists.

````python
from hw4 import cross_validation
````

````python
#### Your code here ####
etas = [0.05, 0.005, 0.0005, 0.00005, 0.000005]
epss = [0.01, 0.001, 0.0001, 0.00001, 0.000001]
folds = 5

random_state = 1
acc_dict = {}
for eta in etas:
    for eps in epss:
        lor = LogisticRegressionGD(eta=eta, eps=eps, random_state=random_state)
        acc = cross_validation(X_training, y_training, folds, lor, random_state=random_state)
        acc_dict[(eta, eps)] = acc
        print(f"Accuracy with eta={eta}, eps={eps}:  {acc:.6f}")

best_params = max(acc_dict, key=acc_dict.get)
best_eta = best_params[0]
best_eps = best_params[1]
````

````python
print(f"Best learning rate (eta): {best_eta}")
print(f"Best epsilon: {best_eps}")

best_acc = acc_dict[best_params]
print(f"Accuracy with the best parameters: {best_acc:.4f}")
````

````python
## your tests here ##

datasets = generate_datasets()

def plot_datasets(datasets):
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))  # 2 rows for two datasets, 3 cols for 3 feature combinations
    titles = ['Feature 1 vs Feature 2', 'Feature 1 vs Feature 3', 'Feature 2 vs Feature 3']
    data_keys = ['dataset_a', 'dataset_b']

    for i, key in enumerate(data_keys):
        features = datasets[f"{key}_features"]
        labels = datasets[f"{key}_labels"]
        for j in range(3):
            ax = axes[i, j]
            f1, f2 = j % 3, (j + 1) % 3  # Select features to plot against each other
            ax.scatter(features[labels == 0, f1], features[labels == 0, f2], color='red', label='Class 0')
            ax.scatter(features[labels == 1, f1], features[labels == 1, f2], color='blue', label='Class 1')
            ax.set_xlabel(f'Feature {f1 + 1}')
            ax.set_ylabel(f'Feature {f2 + 1}')
            ax.set_title(f'{key.upper()} - {titles[j]}')
            ax.legend()

    plt.tight_layout()
    plt.show()

````

````python
plot_datasets(datasets)
````

````python
from mpl_toolkits.mplot3d import Axes3D

def plot_comparative_3d_scatter(datasets, title='Comparative 3D Scatter Plot'):

    fig = plt.figure(figsize=(14, 7))
    fig.suptitle(title)

    # Settings for dataset_a
    ax1 = fig.add_subplot(121, projection='3d')
    features_a = datasets['dataset_a_features']
    labels_a = datasets['dataset_a_labels']
    class1_a = features_a[labels_a == 0]
    class2_a = features_a[labels_a == 1]
    ax1.scatter(class1_a[:, 0], class1_a[:, 1], class1_a[:, 2], c='red', marker='o', label='Class 0')
    ax1.scatter(class2_a[:, 0], class2_a[:, 1], class2_a[:, 2], c='blue', marker='^', label='Class 1')
    ax1.set_title('Dataset A')
    ax1.set_xlabel('Feature 1')
    ax1.set_ylabel('Feature 2')
    ax1.set_zlabel('Feature 3')
    ax1.legend()

    # Settings for dataset_b
    ax2 = fig.add_subplot(122, projection='3d')
    features_b = datasets['dataset_b_features']
    labels_b = datasets['dataset_b_labels']
    class1_b = features_b[labels_b == 0]
    class2_b = features_b[labels_b == 1]
    ax2.scatter(class1_b[:, 0], class1_b[:, 1], class1_b[:, 2], c='red', marker='o', label='Class 0')
    ax2.scatter(class2_b[:, 0], class2_b[:, 1], class2_b[:, 2], c='blue', marker='^', label='Class 1')
    ax2.set_title('Dataset B')
    ax2.set_xlabel('Feature 1')
    ax2.set_ylabel('Feature 2')
    ax2.set_zlabel('Feature 3')
    ax2.legend()

    plt.show()
````
